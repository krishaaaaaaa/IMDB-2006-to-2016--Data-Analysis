# 22. Find Unique Values From Genre 

```python
data['Genre'].str.split(',')
```

```python
list1 =[]
for value in data['Genre']:
    list1.append(value.split(','))
list1
```

```python
unique_list=[]
for item in list1:
    for item1 in item:
        if item1 not in unique_list:
            unique_list.append(item1)
unique_list
```
